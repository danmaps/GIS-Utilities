```python
import pandas as pd

# URL of the webpage containing the table
url = "https://dot.ca.gov/programs/traffic-operations/legal-truck-access/truck-stops-removed#california"

# Read the HTML tables from the webpage
tables = pd.read_html(url)

# Assuming the table you want is the first one
df = tables[0]

df.head()
```

```python
# add a new column that contains the value from "address" column, " ", and "city" column
df["FULL_ADDRESS"] = df["ADDRESS"] + ", " + df["CITY"]+ ", CA"
df.head()
```

```python
# convert this dataframe to a csv
df.to_csv("truck_stops.csv", index=False)
```
